```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Loading training data
train_data = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRTK2NvcndgPX41Czu6Ft2Ho_nE-z50BgTqdzwFW0rsJ2nvyNLe2DoIg1COzUbgw80oaRBjfy5-WtFk/pub?output=csv')

# Droping rows with NaN in y column
train_data = train_data.dropna(subset=['y'])

X_train = train_data[['x']].astype(float)
y_train = train_data[['y']].astype(float)

# Training linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Loading test data
test_data = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRyvZ7lknwiSghK9aen1SaTEYoN3JS40rrGLpcyrsVZy1tB2T4gn6Y3-cdzPUFCPMmmqREWefW3kl4_/pub?output=csv')

X_test = test_data[['x']].astype(float)

# Making predictions on test data
y_pred = model.predict(X_test)

print(y_pred)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load and prepare data
train_data = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRTK2NvcndgPX41Czu6Ft2Ho_nE-z50BgTqdzwFW0rsJ2nvyNLe2DoIg1COzUbgw80oaRBjfy5-WtFk/pub?output=csv')
train_data = train_data.dropna(subset=['y'])

X_train = train_data[['x']].astype(float)
y_train = train_data[['y']].astype(float)

test_data = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRyvZ7lknwiSghK9aen1SaTEYoN3JS40rrGLpcyrsVZy1tB2T4gn6Y3-cdzPUFCPMmmqREWefW3kl4_/pub?output=csv')
X_test = test_data[['x']].astype(float)
y_test = test_data[['y']].astype(float)


# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Convert arrays to 1D for DataFrame constructor
y_test = y_test.squeeze()
y_pred = y_pred.squeeze()

# Output predictions and metrics
results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(results)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

metrics = pd.DataFrame({'MSE': [mse], 'R2': [r2]})
print(metrics)
```
